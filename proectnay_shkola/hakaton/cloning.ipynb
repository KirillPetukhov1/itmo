```python
labels = []
for i in labels_old:
    labels.append([i[0], i[1]])
```

```python
blacklist = '4343 4487 4518 4791 4792 5767 6193 6218 6250 6342 6526 6644 6779 6786 7076 7077 7262 6029 6065 6252 6288 6389 6454 6514 6521 6571 6599 6707 7136 14473'.split()
```

```python
for _ in range(20):    
    for i in range(len(df)):
        if df['idx'][i] in blacklist:
            if random.choice([0, 1, 1, 1, 1]):
                continue
        label_count = list(filter(lambda x: x[0] == df['label'][i], labels))[0][1]
        
        if label_count*2 <= 49:
            total_count = 49//label_count - 1
            for j in range(len(labels)):
                if labels[j][0] ==  df['label'][i]:
                    labels[j][1] += total_count
            for j in range(total_count):
                new_row = df.iloc[i].copy()
                new_row['idx'] = f'{new_row['idx']}_{j+1}'
                df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
                
        elif label_count <= 49:
            for j in range(len(labels)):
                if labels[j][0] ==  df['label'][i]:
                    labels[j][1] += 1
            new_row = df.iloc[i].copy()
            new_row['idx'] = f'{new_row['idx']}_new'
            df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
            
```

```python
def clean_text_spaces(df):
    """
    Удаляет все двойные пробелы, переносы строк и табуляции из столбца 'text' DataFrame
    
    Parameters:
    df (pd.DataFrame): Входной DataFrame
    
    Returns:
    pd.DataFrame: DataFrame с очищенным столбцом 'text'
    """
    # Создаем копию DataFrame, чтобы не изменять оригинал
    result_df = df.copy()
    
    # Применяем замены к столбцу 'text'
    result_df['text'] = result_df['text'].astype(str).apply(
        lambda x: re.sub(r'\s+', ' ', x).strip()
    )
    
    return result_df
```

```python
def low(df):
    df_copy = df.copy()
    # Заполняем NaN пустой строкой и приводим всё к str, затем в нижний регистр
    df_copy['text'] = df_copy['text'].fillna('').astype(str).str.lower()
    return df_copy
```

```python
df = clean_text_spaces(df)
df = low(df)
df=df.sort_values(by=['label', 'idx']).reset_index(drop=True)
```

```python
df.to_csv('train_hackathon_dataset_cloned.csv')
```
