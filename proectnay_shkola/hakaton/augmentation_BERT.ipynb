```python
def augment_with_masked_lm(base_text, augments_prob=0.5):
    # Если num_augments меньше 0, то num_augments - раз в какое в среднем количество слов происходит замена
    results = []
    
    i_start = 0
    splitted_text = base_text.split('.')
    chunked_text = []
    for i in range(len(splitted_text)):
        if sum(map(lambda x: x.count(' '), splitted_text[i_start:i+1])) > 250:
            chunked_text.append(''.join(splitted_text[i_start:i]))
            i_start = i
            
    if i_start != len(splitted_text) - 1:
        chunked_text.append(''.join(splitted_text[i_start:]))
    
    for text in chunked_text:
        words = text.split()
        i = 0

        while i < len(words):
            # print(f'{i}/{len(words)}')
            new_words = words.copy()
            
            if random.randint(0, 1) < augments_prob:
                new_words[i] = mlm.tokenizer.mask_token
                old_word = words[i]
                masked_text = " ".join(new_words)

                # получаем предсказания модели
                preds = mlm(masked_text, top_k=3)
                if preds[0]['token_str'] == '<' or preds[0]['token_str'] == '>':
                    i+=1
                    # print(old_word, preds[0]['token_str'])
                    continue
                if preds[0]['token_str'] == old_word:
                    words = fix_tags(preds[1]['sequence']).split()
                    i+=1
                    continue
                words = fix_tags(preds[0]['sequence']).split()
                # print(old_word, preds[0]['token_str'])
            
            i+=1
            
        results.append(' '.join(words))

    return ' '.join(results)
```
