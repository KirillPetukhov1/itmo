# Полиномиальные признаки

Полиномиальные признаки отражают нелинейные зависимости путем создания квадратов или членов более высокого порядка существующих признаков.

#### Булевые признаки

Булевы признаки — это двоичные индикаторы, которые сигнализируют о наличии или отсутствии определенных условий.

#### Binning (Дискретизация)

Биннинг преобразует непрерывные переменные в категориальные. Это помогает выявить нелинейные зависимости и снизить влияние выбросов.
